**Bar chart for quarters of an year**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


#climate_1 = climate_travel

#climate_1=climate_travel[(climate_travel['Quarter']=='2021Q1') & (climate_travel['YY']=='2021')]
climate_1=climate_travel[(climate_travel['YY']=='2022')]
#
labels = climate_1['Quarter']
Temparature = climate_1['Temp_Value']
Rainfall = climate_1['Rain_Value']
sunshine=climate_1['Sunshine_Value']
travel=climate_1['VALUE']

x = np.arange(len(labels))

width = 0.25
#width = 0.5

fig, ax = plt.subplots(figsize=(8, 6))

rects1 = ax.bar(x - width, Temparature, width, label='Temperature [Degrees]')
rects2 = ax.bar(x, Rainfall, width, label='Rainfall [Millimetres]')
rects3 = ax.bar(x + width, sunshine, width, label='Sunshine  [Hours]')
rects4 = ax.bar(x + width + width, travel, width, label='Travel [Trips]')


ax.set_xticks(x)
ax.set_xticklabels(labels)


ax.set_ylabel('Values')
ax.set_xlabel('Quarter')
ax.set_title('Weather vs Travel')
ax.legend()

plt.show()

#climate_travel
```

**Function to pass year value and draw chart for quarters of that year**

```python
import matplotlib.pyplot as plt
def draw_chart(v_year):
    #climate_1 = climate_travel

    #climate_1=climate_travel[(climate_travel['Quarter']=='2021Q1') & (climate_travel['YY']=='2021')]
    climate_1=climate_travel[(climate_travel['YY']=='2022')]
    climate_1=climate_travel[(climate_travel['YY']==v_year)]

    #
    labels = climate_1['Quarter']
    Temparature = climate_1['Temp_Value']
    Rainfall = climate_1['Rain_Value']
    sunshine=climate_1['Sunshine_Value']
    travel=climate_1['VALUE']

    x = np.arange(len(labels))

    width = 0.25
    #width = 0.5

    fig, ax = plt.subplots(figsize=(8, 6))

    rects1 = ax.bar(x - width, Temparature, width, label='Temperature [Degrees]')
    rects2 = ax.bar(x, Rainfall, width, label='Rainfall [Millimetres]')
    rects3 = ax.bar(x + width, sunshine, width, label='Sunshine  [Hours]')
    rects4 = ax.bar(x + width + width, travel, width, label='Travel [Trips]')


    ax.set_xticks(x)
    ax.set_xticklabels(labels)


    ax.set_ylabel('Values')
    ax.set_xlabel('Quarter')
    ax.set_title('Weather vs Travel')
    ax.legend()

plt.show()

#climate_travel
```

```python
draw_chart('2019')
draw_chart('2020')
draw_chart('2021')
draw_chart('2022')
```
